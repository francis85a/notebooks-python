## Diccionarios

#### Basic Dictionaries Operations

Se crea un diccionario y se pide que represente el valor de spam y luego que represente el diccionario entero

```python
D = {'spam': 2, 'ham': 1, 'eggs': 3}
print(D['spam'])
print (D)
```

len recorre el diccionario y devuelve la cantidad de objetos del diccionario, in comprueba si ese objeto esta en el diccionario y list(.keys) crea una lista con las claves definidas en el diccionario

```python
print (len(D))
print('ham' in D)
list(D.keys())
```

#### Changing Dictionaries in Place

Cambia el valor de la clave ham con una lista

```python
print(D)
D['ham'] = ['grill', 'bake', 'fry']
print(D)
```

"Del" elimina la clave y el valor del diccionario que le indiques

```python
del D['eggs']
print(D)
```

Aade una nueva clave con su valor al diccionario

```python
D['brunch'] = 'Bacon'
print(D)
```

#### More Dictionary Methods

"list(D.values)" devuelve los valores que estan definidos en el diccionario en una lista y "list(D.items)" devuelve las claves y valores definidos en el diccionario en una lista

```python
D = {'spam': 2, 'ham': 1, 'eggs': 3}
print(list(D.values()))
list(D.items())
```

.get comprueba si esta esa clave en el diccionario si esta te devuelve su valor y si no devuelve None

```python
print(D.get('spam'))
print(D.get('toast'))
D.get('toast', 88)
```

.update aade los objetos de un diccionario en otro como si los concatenara

```python
print(D)
D2 = {'toast': 4, 'muffin': 5}
D.update(D2)
print (D)
```

.pop elimina la clave del diccionario que le indiques y devuelve su valor

```python
print(D)
print(D.pop('muffin'))
print(D.pop('toast'))
print(D)
```

#### Example: Movie Database

crea un diccionario y segun el ao que pongas te devuelve la pelicula que se estreno ese ao

```python
table = {'1975': 'Holy Grail', '1979': 'Life of Brian', '1983': 'The Meaning of Life'}

year = '1983'
movie = table[year]
movie
```

hace un bucle que recorra el diccionario y devuelva cada ao con la pelicula

```python
for year in table:
    print(year + '\t' + table[year])
```

Crea una lista con las claves y valores del diccionario y luego hace una funcion que compruebe si el titulo y el ao de la pelicula estan en el diccionario y si son iguales que el ao que se pone que devuelva el titulo de la pelicula

```python
table = {'Holy Grail': '1975', 'Life of Brian': '1979', 'The Meaning of Life': '1983'}
print(table['Holy Grail'])

print(list(table.items()))
print([title for (title, year) in table.items() if year == '1975'])
```

```python
K = 'Holy Grail'
print(table[K])

V = '1975'
print([key for (key, value) in table.items() if value == V])
print([key for key in table.keys() if table[key] == V])
```

Puedes crear un diccionario para hacer matrices y darle valores a Letras como si fueran las cordenadas y que tomen el valor del diccionario

```python
Matrix = {}
Matrix[(2, 3, 4)] = 88
Matrix[(7, 8, 9)] = 99

X = 2; Y = 3; Z = 4
print(Matrix[(X, Y, Z)])
Matrix
```

```python
if (2, 3, 6) in Matrix:
    print(Matrix[(2, 3, 6)])
else:
    print(0)

try:
    print(Matrix[(2, 3, 6)])
except KeyError:
    print(0)

print(Matrix.get((2, 3, 4), 0))
Matrix.get((2, 3, 6), 0)
```
