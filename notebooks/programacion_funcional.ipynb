# Herramientas Programacion Funcional

### Funciones anonimas: lambda

la funcion lambda se llama despues pero por su funcion y no por su nombre por eso se le llama funcion anonima

```python
def func(x, y, z):
    return x + y + z
func(2, 3, 4)
```

```python
f = lambda x, y ,z: x + y + z
f(2, 3, 4)
```

### porque usar lambda?

```python
L = [lambda x: x ** 2,
     lambda x: x ** 3,
     lambda x: x ** 4]
for f in L: print(f(2))

print (L[0](3))
```

```python
f = (lambda a = "fee", b="fie", c="foe": a + b + c)
f("wee")
```

```python
def knights():
    title = 'Sir'
    action = (lambda x: title + ' ' + x)
    return action

act = knights()
msg = act('robin')
msg
```

Las expresiones lambda introducen un ambito local parecido a una definicion anidad 'def' mediante la regla LEGB

### Funciones mapeadas: map

```python
counters = [1, 2, 3, 4]
updated = []
for x in counters:
    updated.append(x + 10)
updated
```

```python
def inc(x): return x + 10

list (map(inc, counters))
```

map llama a inc en cada objeto de la lista y los recolecta todos y devuelve los valores en una nueva lista

```python
lista = list(map((lambda x: x + 3), counters))
print(lista)

counters
```

##### Secuencias con multiple argumento

```python
pow(3, 4)

list(map(pow, [1, 2, 3], [2, 3, 4]))
```

##### Map compresiones similares

```python
list(map(inc, [1, 2, 3, 4]))

[inc(x) for x in [1, 2, 3, 4]]
```

### Seleccionar objetos en iterables: filter

```python
a = list(range(-5, 5))
print(a)

b = list(filter((lambda x: x > 0), range(-5, 5)))
print(b)

res = []
for x in range(-5, 5):
    if x > 0:
        res.append(x)
res
```

```python
[x for x in range(-5, 5) if x > 0]
```

### Combinar objetos en iterables: reduce

```python
from functools import reduce
reduce((lambda x, y: x + y), [1, 2, 3, 4])
```

```python
L = [1, 2, 3, 4]
res = L[0]
for x in L[1:]:
    res = res + x
res
```
