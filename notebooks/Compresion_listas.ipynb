# Compresiones en Listas

------

```python
L = [1, 2, 3, 4, 5]
for i in range(len(L)):
    L[i] += 10
L
```

La compresion seria la siguiente dandole el valor x a cada elemento de la secuencia

```python
L = [x + 10 for x in L]
L
```

Es lo mismo pero seria sin compresion

```python
res = []
for x in L:
    res.append(x + 10)
res
```

----------

### Compresion sintactica de listas extendidas

##### Filter clauses: if

```python
L = [1, 7, 8, 9]
M = [x for x in L if x < 5]
M
```

##### Bucles anidados: for

```python
[x + y for x in 'abc' for y in 'lmn']
```

es lo mismo que lo siguiente:

```python
res = []
for x in 'abc':
    for y in 'lmn':
        res.append(x + y)
res
```

##### Tuples

Se pueden usar las compresiones de listas para convertir tuples

```python
T = (1, 2, 3, 4, 5)
L = [x + 20 for x in T]
L
```

##### Range

```python
L = [x for x in range(0,9)]
L
```

### Compresiones de diccionarios

Ejemplo 1 --
zip(*iterables)

```python
x = [1, 2, 3]
y = [4, 5, 6]
zipped = zip(x, y)
list(zipped)
```

```python
L = [x for x in range(0,4)]
M = [x for x in range(5,10)]
[elemento for elemento in zip(L,M)]
```

```python
list(zip(['a', 'b', 'c'], [1, 2, 3]))

D = dict(zip(['a', 'b', 'c'], [1, 2, 3]))
D

D = {k: v for (k, v) in zip(['a', 'b', 'c'], [1, 2, 3])}
D
```

Ejemplo 2

```python
D = dict.fromkeys(['a', 'b', 'c'], 0)		
D

D = { k:0 for k in ['a', 'b', 'c'] }			
D
```
