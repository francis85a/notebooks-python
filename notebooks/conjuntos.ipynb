### Operaciones sobre conjuntos

```python
S1 = {1, 2, 3, 4}
S1 & {1,3}
# Hace una intersección y devuelve los elementos comunes
```

```python
{1, 5, 3, 6} | S1
# Hace una unión y devuelve todos los elementos sin repetir
```

```python
S1 - {1, 3, 4}
# Hace una diferencia y devuelve los elementos que no están en el segundo conjunto
```

```python
S1 > {1, 3}
# Comprueba si el primer conjunto es un superconjunto del segundo
```

{} es un diccionario
los conjuntos deben ser creados con la funcion 'set' y printarlos de la misma forma

```python
S1 - {1, 2, 3, 4}
```

```python
type({})
```

```python
S = set()
S.add(1.23)
S
# se crea un conjunto vacío y se le añade un elemento
```

### Vistas diccionarios y sets

La vista objeto devuelta por keys es como un conjunto y soporta las operaciones de los conjuntos

```python
D = dict(a=1, b=2, c=3)
K = D.keys()
V = D.values()
del D['b']
D
```

```python
K, V
```

```python
K | {'x': 4}
```

el metodo keys permite hacer union en el diccionario creado como si fuera un conjunto

```python
V & {'x': 4}
```

El metodo valores no permite hacer intersecciones
