#### Objetos Mutables

##### Listas

```python
L = [2, 3, 4]	    # Las listas son mutables
M = L			    # creas una referecia de la lista con un string
L[0] = 24 		    # sustituyes un elemento de la lista

print("L = ", L)    # [24, 3, 4]
print("M = ", M)    # [24, 3, 4]
```

Ya que las listas son mutables puedes hacer un string referenciando a la lista y luego puedes modificar un caracter de la lista, entonces la lista se modifica y aunque el string este encima de la modificaciÃ³n se representa el cambio en el resultado

```python
L = [2, 3, 4]
M = L[:]		    # crea una copia de L
L[0] = 24

print("L = ", L)    # [24, 3, 4]

print("M = ", M)    # [2, 3, 4]  M no se cambia
```

En este caso el string esta haciendo una copia de la lista y por eso al reprensetarse sale la lista principal y no la modificada

#### Compartir referencias y igualdad

 - Operador "==" iguala valores
 - Operador "is" identifica objetos

```python
L = [1, 2, 3]
M = L 			    # M referencia a L

L == M 			    # True: iguala valores=> operador igualdad de valores
L is M 			    # True: operador identidad de objetos: compara las referencias (los punteros)
```

```python
L = [1, 2, 3]
M = [1, 2, 3]	    # M y L son dos listas diferentes

print(
    "L == M",
    L == M 			# True: compara valores
)

print(
    "L is M", 	    # False: operador identidad de objetos: compara las referencias (los punteros)
    L is M
)
```

Aunque L y M son listas con los mismos valores no es la misma lista por eso al igualar valores dice que es Verdadero pero al comparar si son el mismo objeto es falso
