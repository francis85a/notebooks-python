### Recorrido de los elementos

un diccionario no es una secuencia pero se puede adaptar para recorrerlo con un bucle for

podemos acceder a las claves mediante "keys()" y a los valores mediante "values()"

```python
codigos = {}
codigos["Programacion"] = "MP0485"
codigos["Sistemas informaticos"] = "MP0483"
codigos["Contornos de desenvolvemento"] = "MP0487"
codigos["Bases de datos"] = "MP0484"
codigos["Linguaxes de marcas"] = "MP0373"
codigos["Formacion e orientacion laboral"] = "MP0617"

for clave in codigos: # se llama al método __iter__() del diccionario de forma implícita
    print(clave, codigos[clave])

for clave in codigos.keys():
    print(clave, codigos[clave])

for valor in codigos.values():
    print(valor)

for clave, valor in codigos.items(): # Devuelve una tupla y desempaqueta
    print(clave, valor)

for tupla in codigos.items(): # Devuelve una tupla
    print(tupla)  
```

### Modificacion de los elementos

pueden modificarse los valores mediante asignacion aunque se este iterando

```python
precios_frutas = {
                "manzana": 2.35,
                "pera": 3.05,
                "naranja": 1.75,
                "limon": 1.95,
                "melon": 4.85,
                "sandia": 3.45,
                "platano": 2.95,
                "cereza": 5.90,
                "uva": 4.15,
                "fresa": 4.25,
                }

print(precios_frutas)

precios_frutas["fresa"] = 6.54

print(precios_frutas)

for fruta in precios_frutas:
    precios_frutas[fruta] = round(precios_frutas[fruta] * 1.10, 2) # incrementamos el precio de la fruta en un 10%

print(precios_frutas)
```

### Eliminación de un elemento

se utiliza "del" seguida del elemento que se quiere eliminar

### Funciones que pueden aplicarse a los diccionarios

len(diccionario) devuelve el numero de elementos del diccionario

```python
telefonos = {
    "Juan": 123456789,
    "Pedro": 987654321,
    "Ana": 654987321,
    "Maria": 321654987,
    }
print(len(telefonos))
```

list(diccionario) devuelve una lista con las claves del diccionario

```python
personas = list(telefonos)

print(personas)
print(telefonos.keys())
```

in y not in se pueden usar para saber si una clave esta o no en el diccionario

```python
print("Juan" in telefonos)
print("Andrea" in telefonos)
```

### Metodos de la clase dict

clear() elimina todos los elementos del diccionario

la diferencia con del es que clear() no elimina el diccionario, si no que lo deja vacio

```python
precios_frutas = {
    "manzana": 2.35,
    "pera": 3.05,
    "naranja": 1.75,
    "limon": 1.95,
    "melon": 4.85,
    "sandia": 3.45,
    "platano": 2.95,
    "cereza": 5.90,
    "uva": 4.15,
    "fresa": 4.25,
    }
print(precios_frutas)
precios_frutas.clear()
print(precios_frutas)
```

pop(clave) elimina el elemento con la clave indicada y devuelve su valor

```python
precios_frutas = {
                "manzana": 2.35,
                "pera": 3.05,
                "naranja": 1.75,
                "limon": 1.95,
                "melon": 4.85,
                "sandia": 3.45,
                "platano": 2.95,
                "cereza": 5.90,
                "uva": 4.15,
                "fresa": 4.25,
                }

precio = precios_frutas.pop("manzana") # Elimina el elemento con clave "manzana" y devuelve su valor

print(precio)
print(precios_frutas)
```

popitem() elimina el ultimo elemento y devuelve su clave y valor

```python
print(precios_frutas.popitem())
```

update(diccionario) añade los pares clave-valor del diccionario indicado y modifica los que ya estan

```python
diccionario = {
    "hola":"привіт", 
    "adios":"бувай",
    "casa":"дім",
    "perro":"пес",
    "gato":"кіт",
    }
mas_palabras = {
    "coche":"автомобіль",
    "libro":"книга",
    "ordenador":"комп'ютер",
    "moto":"мотоцикл",
    "casa":"будинок",
    }
print(diccionario)
diccionario.update(mas_palabras)
print(diccionario)
```
