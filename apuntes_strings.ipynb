#### Type-Specific Methods

- Find

```python
# busca los caracteres en el string y devuelve la posicion donde se encuentra
s = 'SpamEggsSpam'
s.find('pa')
```

```python
# si hay mas de un caracter igual en el string tienes que especificar donde acaba el primer caracter encontrado para que busque el siguiente
s.find('pa', 3)
```

```python
# Si no existe el caracter devuelve -1
s.find('z')
```

- Replace

```python
# aunque sea inmutable el string se puede remplazar para que devuelva otra cosa distinta
s.replace('Eggs','Bacon')
```

- Mayusculas

```python
# tambien aunque sea inmutable se puede modificar para que salga en mayusculas
s.upper()
```

```python
# podemos comprobar si una letra del string es mayuscula o no
print("s is alpha", s.isalpha())
print("s is digit", s.isdigit())
```

- Split

```python
# se pueden dividir las palabras para separarlas
s = 'spams-spam-eggs-spam-bacon'
s.split('-')
```

- Rstrip

```python
# elimina espacios en blanco a la derecha
line = 'aaa\t,bbb\t,\nccccc,\tdd\n'
print(line)
```

```python
line.rstrip()
```

```python
line.lstrip()
```

```python
# combinamos las dos operaciones
line.rstrip().split(',')
```

- Formatting

```python
# tambien se puede hacer sustitución dandole distinto formato al string
# expresion de formato
'%s,eggs, and %s' % ('spam','SPAM!')
```

```python
# metodo de formato
'{}, eggs, and {}'.format('spam', 'SPAM!')
```

```python
s = 'SpamEggsSpam'
f'{s}'
```

```python
'{0}'.format(s)
```
